```
from math import sqrt

#Cost

cost = [10, 12, 14, 15, 16, 16, 17, 18.5, 19, 19.5]
costmean = sum (cost) / float(len(cost))
costsum = sum(cost)
costmin = min(cost)
costmax = max(cost)
costmode = max(set(cost), key = cost.count)

#Finding the Cost Median

cost.sort()
length = len(cost)
if (length % 2 == 0):
    costmedian = (cost[(length)//2] + cost[(length)//2-1]) / 2
else:
    costmedian = cost[(length-1)//2]
    
#Standard Error of Cost variables from Mean

costhistderror = []
costlowstderror = []

for i in cost:
  if i >= costmean:
    costhistderror.append(i)
  elif i <= costmean:
      costlowstderror.append(i)
      
#Standard Deviation of Cost

costvariance = sum((i-costmean) **2 for i in cost) / (len(cost))
coststdev = sqrt(costvariance)
      
#Time

time = [5, 7, 8, 9, 10, 10, 11, 11.5, 12, 13]
timemean = sum(time) / float(len(time))
timesum = sum(time)
timemin = min(time)
timemax = max(time)
timemode = max(set(time), key = time.count)

#Finding the Time Median

time.sort()
length = len(time)
if(length % 2 == 0):
  timemedian = (time[length//2] + time[length//2-1]) /2
else:
  timemedian = (time[(length-1)//2])
  
#Standard Error of Time variables from Mean

timehistderror = []
timelowstderror = []

for i in time:
  if i >= timemean:
    timehistderror.append(i)
  elif i <= timemean:
      timelowstderror.append(i)
      
#Standard Deviation of Time

timevariance = sum((i-timemean) **2 for i in time) / (len(time))
timestdev = sqrt(timevariance)

#Units

units = [19, 21, 22, 23, 24, 24, 25, 26, 27, 28]
unitsmean = sum(units) / float(len(units))
unitssum = sum(units)
unitsmin = min(units)
unitsmax = max(units)
unitsmode = max(set(units), key = units.count)

#Finding the Units Median

units.sort()
length = len(units)
if (length % 2 == 0):
    unitsmedian = (units[(length)//2] + units[(length)//2-1]) / 2
else:
    unitsmedian = units[(length-1)//2]
    
#Standard Error of Units variables from Mean

unitshistderror = []
unitslowstderror = []

for i in units:
  if i >= unitsmean:
    unitshistderror.append(i)
  elif i <= unitsmean:
      unitslowstderror.append(i)
      
#Standard Deviation of Units

unitsvariance = sum((i-unitsmean) **2 for i in units) / (len(units))
unitsstdev = sqrt(unitsvariance)

#Cost Output

print ('\033[1m', 'For Cost','\033[0m', '\n')
print ('\tSum=',costsum, '\n', '\tMin=',costmin, '\n', '\tMax=', costmax, '\n', '\tMode=', costmode, '\n', '\tMedian=',costmedian, '\n')
print ('\033[1m' ,'Standard Error of Cost variables from Mean:','\033[0m','\n')
print ('\tMean of cost= $',costmean, '\n')
print ('\tValues higher than Mean:',costhistderror, '\n')
print ('\tValues lower than Mean:',costlowstderror, '\n')
print ('\033[1m' ,'Standard Deviation of Cost:', '\033[0m', coststdev, '\n')
print ('\033[1m', ('-' * 75),'\033[0m', '\n')

#Time Output

print ('\033[1m','For Time:','\033[0m','\n')
print ('\tSum=',timesum, '\n', '\tMin=',timemin, '\n', '\tMax=', timemax, '\n', '\tMode=', timemode,'\n', '\tMedian=', timemedian, '\n' )
print ('\033[1m''Standard Error of Time variables from Mean:','\033[0m', '\n')
print ('\tMean of Time=',timemean,'mins', '\n')
print ('\tValues higher than Mean:',timehistderror, '\n')
print ('\tValues lower than Mean:',timelowstderror, '\n')
print ('\033[1m''Standard Deviation of Time:', '\033[0m', timestdev, '\n')
print ('\033[1m', ('-' * 75),'\033[0m', '\n')

#Units Output

print ('\033[1m', 'For Units','\033[0m', '\n')
print ('\tSum=',unitssum, '\n', '\tMin=',unitsmin, '\n', '\tMax=', unitsmax, '\n', '\tMode=', unitsmode, '\n', '\tmedian=',unitsmedian, '\n')
print ('\033[1m' ,'Standard Error of Units variables from Mean:','\033[0m','\n')
print ('\tMean of Units=', unitsmean, '\n')
print ('\tValues higher than Mean:',unitshistderror, '\n')
print ('\tValues lower than Mean:',unitslowstderror, '\n')
print ('\033[1m' ,'Standard Deviation of Units:', '\033[0m', unitsstdev, '\n')
```
